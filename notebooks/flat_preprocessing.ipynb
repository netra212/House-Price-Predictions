## **Basic Cleaning on `Flat` `Dataset`**

```python
import pandas as pd
import numpy as np

# printing all the columns and rows names.
flat = pd.read_csv("/content/drive/MyDrive/DATASET_ML/flats.csv")
flat.sample(5)
```

```python
# shape
flat.shape
```

```python
# info
flat.info()
```

All the features mentioned in the datasets are string data types.

```python
# Check for the duplicates.
flat.duplicated().sum()
```

The given flat datasets does not contains any duplicated values.

```python
# Check for the missing values.
flat.isnull().sum()
```

In the above result, features like `additonalRoom`, `facing`, `nearbyLocations`, `furnishDetails`, features, rating contains more number of missing values as compare to other features such as `society`, `price`, `area`, `areaWithType`, `bedRoom`, `bathroom`, `balcony`, `address`, `flatNum`, `agePossession`, `description`, `property_id`.

In the above datasets, we don't require `link` and `property_id` column. so, Can be drop.

```python
# Dropping `link` and `property_id` column.
flat.drop(columns=['link', 'property_id'], inplace=True)
flat.head(5)
```

Since, In the above dataset, there is a `area` column with values in the **price per square foot**. so, renaming the `area` column with `price_per_sqft` for better readibility.

```python
flat.rename(columns={
    'area':'price_per_sqft',
}, inplace=True)
flat.head(6)
```

Now, Let's work on the explore into the each features only by one, starting from `society` column leaving `property_name` for now.

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# society column.
flat['society'].value_counts()
```

In the above `society` column, there is also a rating attached with that so needs to remove. And all the names of the society needs to convert into lower case letter.

```python
# society column.
flat['society'].value_counts().shape
```

```python
import re

flat['society'] = flat['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip())
```

```python
flat['society'] = flat['society'].str.lower()
flat['society']
```

```python
flat['society'].value_counts().shape
```

Now, If we observed, the shape of the `society` column now reduced from 638 to 611.

```python
flat['society'].shape
```

```python
flat.head()
```

```python
# Now, Let's work into `price` column.
# Workflow: Converting every Lac in term of Crore.
flat['price'].value_counts().sort_values(ascending=False)
```

While observing the above `price` column values, there is also a **Price on request**.

```python
flat[flat['price'] == "Price on Request"]
```

There are all together 11 rows of data where `price` col contains values = `"Price on Request"`.

```python
# Fetching all the values of price columns where values are not equal to "Price on Request"
flat = flat[flat['price'] != "Price on Request"]
flat.head(10)
```

```python
# Converting price value where values = Lacs in terms of crores.

# x = flat['price'].str.split(' ')

def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)

flat['price'] = flat['price'].str.split(' ').apply(treat_price)
```

```python
flat.head(5)
```

```python
# Now, Let's work on the price_per_sqft columns.
flat['price_per_sqft'].value_counts().sort_values(ascending=True)
```

```python
flat['price_per_sqft'] = flat['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',','').str.strip().astype('float')
```

```python
flat.head(5)
```

```python
# Now, Let's work on the bedRoom.
flat['bedRoom'].value_counts()
```

```python
flat['bedRoom'].isnull().sum()
```

```python
# Taking that values which are not null.
flat = flat[~flat['bedRoom'].isnull()]

flat['bedRoom'] = flat['bedRoom'].str.split().str.get(0).str.replace('Bedrooms', '').str.strip().astype('int')
```

```python
flat.head(5)
```

```python
# Now, Let's work into the bathrooms columns.
flat['bathroom'].value_counts()
```

```python
flat['bathroom'].isnull().sum()
```

```python
flat['bathroom'] = flat['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
flat['bathroom'].value_counts()
```

```python
flat.head(5)
```

```python
# Now, Let's work into the balcony.
flat['balcony'].isnull().sum()
```

```python
flat['balcony'].value_counts()
```

```python
flat['balcony'] = flat['balcony'].str.split(' ').str.get(0).str.replace("No", "0")
```

```python
flat.head(5)
```

```python
# Now, Let's work into the additionalRoom column.
flat['additionalRoom'].value_counts()
```

```python
# Check the shape of the additionalRoom
flat['additionalRoom'].value_counts().shape
```

```python
flat['additionalRoom'].isnull().sum()
```

```python
flat['additionalRoom']
```

```python
# Filling NaN values with `not available` to enhance the readibility.
flat['additionalRoom'].fillna('not available', inplace=True)
```

```python
flat['additionalRoom'].isnull().sum()
```

```python
flat['additionalRoom'] = flat['additionalRoom'].str.lower()
```

```python
flat.head(5)
```

```python
# Now, Let's work into the floorNum cols.
flat['floorNum'].value_counts().sort_values(ascending=True)
```

```python
flat['floorNum'].isnull().sum()
```

```python
flat[flat['floorNum'].isnull()]
```

```python
flat['floorNum'] = flat['floorNum'].str.split(' ').str.get(0).replace('Ground', '0').str.replace('Basement', '-1').str.replace('Lower', '0').str.extract(r'(\d+)')
```

```python
flat.head(5)
```

```python
# Now, Let's work into the `facing` columns.
flat['facing'].value_counts()
```

```python
flat['facing'].isnull().sum()
```

```python
flat['facing']
```

```python
flat['facing'].fillna('NA', inplace=True)
```

```python
flat.head(5)
```

```python
flat['facing'].value_counts().sort_values(ascending=True)
```

```python
flat.head(5)
```

```python
# Let's calculate the area of the each flat and insert into any place within the dataset.

flat.insert(loc=4,column='area',value=round((flat['price']*10000000)/flat['price_per_sqft']))
```

```python
flat.head(5)
```

```python
# Now, Adding property_type column in the dataset.
flat.insert(loc=1, column='property_type', value='flat')
```

```python
flat.head(6)
```
